Data

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
temp=pd.DataFrame(columns=["max","min"])
temp["max"]=[39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
temp["min"]=[21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
temp
```

```python
temp.plot(kind="line",marker='o',)
plt.xlabel("Month")
plt.ylabel("Temperature $^o$C")
```
